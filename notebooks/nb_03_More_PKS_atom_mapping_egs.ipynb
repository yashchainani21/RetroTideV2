In the previous notebook, we developed an atom-mapping algorithm to track which carbon atoms are being added and/ or transformed by each module in a chimeric type I PKS. Here, we deploy this atom mapping algorithm for a couple of examples

```python
from retrotide import retrotide, structureDB
```

```python
import bcs
```

```python
from collections import OrderedDict
```

```python
### Quick reminder on how to create a bcs cluster object
loading_module_domains = {bcs.AT: bcs.AT(active=True, substrate='Acetyl-CoA')}
loading_module = bcs.Module(domains = loading_module_domains, loading=True)

module1_domains = {bcs.AT: bcs.AT(active=True, substrate='Malonyl-CoA'),
                   bcs.KR: bcs.KR(active=True, type='B'),
                   bcs.DH: bcs.DH(active=True)}

module1 = bcs.Module(domains = module1_domains)
moduleStructure = module1.computeProduct()

cluster = bcs.Cluster(modules=[loading_module, module1])
cluster.computeProduct(structureDB)
```

#### Visualizing atom maps for 4-hydroxybutyric acid

```python
loading_AT_domain = {bcs.AT: bcs.AT(active = True, substrate = "Hydroxymalonyl-CoA")}
loading_module = bcs.Module(domains = loading_AT_domain, loading = True)

mod1_AT_domain = bcs.AT(active = True, substrate = "Malonyl-CoA")
mod1_KR_domain = bcs.KR(active = True, type = "B")
mod1_DH_domain = bcs.DH(active = True)
mod1_ER_domain = bcs.ER(active = True)

# create an ordered dict of domains (has to be ordered to escape warnings)
domains_dict = OrderedDict({bcs.AT: mod1_AT_domain,
                       bcs.KR: mod1_KR_domain,
                       bcs.DH: mod1_DH_domain,
                       bcs.ER: mod1_ER_domain})

module1 = bcs.Module(domains = domains_dict, loading = False)

cluster = bcs.Cluster(modules = [loading_module, module1])
```

```python
hydroxybutyric_acid = cluster.computeProduct(structureDB)
hydroxybutyric_acid
```

Running the `.create_atom_maps` method on the instantiated `cluster` object with `structureDB` as an input argument generates a list of atom-mapped PKS products:

```python
atom_mapped_mols_list = cluster.create_atom_maps(structureDB)
atom_mapped_mols_list
```

The first element in this list is the atom-mapped PKS product generated by the loading module:

```python
atom_mapped_mols_list[0]
```

Meanwhile, the second element in this list is the atom-mapped PKS product generated by the first extension module:

```python
atom_mapped_mols_list[1]
```

We can then use the `.visualize_atom_maps` method to view all the atom-mapped products from each PKS module as an RDKit mols grid:

```python
cluster.visualize_atom_maps(atom_mapped_mols_list)
```

#### Visualizing atom maps for 5-ketohexanoic acid

```python
### create loading module
loading_AT_domain = bcs.AT(active = True, substrate = "Malonyl-CoA")
LM_domains_dict = OrderedDict({bcs.AT: loading_AT_domain})
loading_module = bcs.Module(domains = LM_domains_dict, loading = True)

### create module 1
mod1_AT_domain = bcs.AT(active = True, substrate = "Malonyl-CoA")
mod1_domains_dict = OrderedDict({bcs.AT: mod1_AT_domain})
module1 = bcs.Module(domains = mod1_domains_dict, loading = False)

### create module 2
mod2_AT_domain = bcs.AT(active = True, substrate = "Malonyl-CoA")
mod2_KR_domain = bcs.KR(active = True, type = "B")
mod2_DH_domain = bcs.DH(active = True)
mod2_ER_domain = bcs.ER(active = True)

mod2_domains_dict = OrderedDict({bcs.AT: mod2_AT_domain,
                                 bcs.KR: mod2_KR_domain,
                                 bcs.DH: mod2_DH_domain,
                                 bcs.ER: mod2_ER_domain})

module2 = bcs.Module(domains = mod2_domains_dict, loading = False)

### finally, create PKS cluster
cluster = bcs.Cluster(modules = [loading_module, module1, module2])
```

```python
PKS_product = cluster.computeProduct(structureDB)
PKS_product
```

```python
atom_mapped_mols_list = cluster.create_atom_maps(structureDB)
atom_mapped_mols_list
```

```python
cluster.visualize_atom_maps(atom_mapped_mols_list)
```

#### Visualizing atom maps for cryptofolione

```python
### create loading module
loading_module_AT_domain = bcs.AT(active = True, substrate = "cin")
LM_domains_dict = OrderedDict({bcs.AT: loading_module_AT_domain})

loading_module = bcs.Module(domains = LM_domains_dict, loading = True)

### create extension module 1
mod1_AT_domain = bcs.AT(active = True, substrate = "Malonyl-CoA")
mod1_KR_domain = bcs.KR(active = True, type = "B")
mod1_domains_dict = OrderedDict({bcs.AT: mod1_AT_domain,
                                bcs.KR: mod1_KR_domain})

module1 = bcs.Module(domains = mod1_domains_dict, loading = False)

### create extension module 2
mod2_AT_domain = bcs.AT(active = True, substrate = "Malonyl-CoA")
mod2_KR_domain = bcs.KR(active = True, type = "B")
mod2_domains_dict = OrderedDict({bcs.AT: mod2_AT_domain,
                                 bcs.KR: mod2_KR_domain})

module2 = bcs.Module(domains = mod2_domains_dict, loading = False)

### create extension module 3
mod3_AT_domain = bcs.AT(active = True, substrate = "Malonyl-CoA")
mod3_KR_domain = bcs.KR(active = True, type = "B")
mod3_DH_domain = bcs.DH(active = True)
mod3_domains_dict = OrderedDict({bcs.AT: mod3_AT_domain,
                                 bcs.KR: mod3_KR_domain,
                                 bcs.DH: mod3_DH_domain})

module3 = bcs.Module(domains = mod3_domains_dict, loading = False)

### create extension module 4
mod4_AT_domain = bcs.AT(active = True, substrate = "Malonyl-CoA")
mod4_KR_domain = bcs.KR(active = True, type = "B")
mod4_domains_dict = OrderedDict({bcs.AT: mod4_AT_domain,
                                 bcs.KR: mod4_KR_domain})

module4 = bcs.Module(domains = mod4_domains_dict, loading = False)

### create extension module 5
mod5_AT_domain = bcs.AT(active = True, substrate = "Malonyl-CoA")
mod5_KR_domain = bcs.KR(active = True, type = "B")
mod5_DH_domain = bcs.DH(active = True)
mod5_domains_dict = OrderedDict({bcs.AT: mod5_AT_domain,
                                 bcs.KR: mod5_KR_domain,
                                 bcs.DH: mod5_DH_domain})

module5 = bcs.Module(domains = mod5_domains_dict, loading = False)
```

```python
cluster = bcs.Cluster(modules = [loading_module, module1, module2, module3, module4, module5])
cryptofolione = cluster.computeProduct(structureDB)
cryptofolione
```

```python
cryptofolione_atom_maps = cluster.create_atom_maps(structureDB)
cryptofolione_atom_maps
```

```python
cluster.visualize_atom_maps(cryptofolione_atom_maps)
```
